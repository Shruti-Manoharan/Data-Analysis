## 8 Calculate the average duration of songs in minutes (convert 'duration_ms' to minutes).

```python
df['duration_minutes'] = df['duration_ms'] / 60000
avg = df['duration_minutes'].mean()
print(f"The average duration of songs is {avg:.2f} minutes.")
```

## 9 Count the number of songs released each year and plot it using a line plot.

```python
songs_per_year = df['year'].value_counts().sort_index()
songs_per_year
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(5, 4))
sns.lineplot(data=songs_per_year, marker='o')
plt.title('Number of Songs Released Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.grid(True)
plt.show()
```

## 10 Calculate the total duration of songs for each genre and plot it using a bar chart.

```python
dur_per_genre = df.groupby('playlist_genre')['duration_ms'].sum().sort_index()
total_hours = (dur_per_genre) / (1000*60*60)
total_hours.reset_index()
```

```python
plt.figure(figsize=(5, 4))
sns.barplot(x=total_hours.index , y=total_hours.values, palette='tab10')
plt.title('Duration of songs for each genre')
plt.xlabel('Genre')
plt.ylabel('Duration in hours')
plt.xticks(rotation=45)
plt.show()
```

## 11 Create a scatter plot of danceability versus energy, colored by genre.

```python
plt.figure(figsize=(5, 4))
sns.scatterplot(data=df, x='danceability', y='energy', hue='playlist_genre', palette='tab10')
plt.title('Scatter plot of danceability versus energy')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

## 12 Create a column 'tempo_category' that categorizes songs as 'Slow', 'Moderate',
or 'Fast' based on their tempo (Slow: < 90 BPM, Moderate: 90-140 BPM, Fast: > 140 BPM).

```python
bins = [0, 90, 140, float('inf')]
labels = ['Slow', 'Moderate', 'Fast']
df['tempo_category'] = pd.cut(df['tempo'], bins=bins, labels=labels)
df[['track_name', 'tempo', 'tempo_category']]
```

```python
df['tempo_category'].value_counts()
```

## 13 Find the artist with the highest average popularity across all their songs.

```python
artist_avg_popularity = df.groupby('track_artist')['track_popularity'].mean()
highest_avg_popularity_artist = artist_avg_popularity.idxmax()
highest_avg_popularity = artist_avg_popularity.max()
print(f'The artist with the highest average popularity is {highest_avg_popularity_artist} with an average popularity of {highest_avg_popularity:.2f}.')
```

## 14  Identify duplicate songs based on 'name' and 'artist'. How many duplicate songs are there?

```python
duplicates = df[df.duplicated(subset=['track_name', 'track_artist'], keep=False)]
num_duplicates = duplicates.shape[0] - duplicates['track_id'].nunique()
print(f'There are {num_duplicates} duplicate songs.')
```

## 18  Create a FacetGrid of histograms showing the distribution of 'acousticness' for each 'tempo_category'.

```python
def categorize_tempo(tempo):
    if tempo < 90:
        return 'Slow'
    elif tempo <= 140:
        return 'Moderate'
    else:
        return 'Fast'

df['tempo_category'] = df['tempo'].apply(categorize_tempo)
g = sns.FacetGrid(df, col='tempo_category', margin_titles=True)
g.map(plt.hist, 'acousticness', bins=10, color='skyblue', edgecolor='black')
g.set_axis_labels('Acousticness', 'Frequency')
g.fig.suptitle('Distribution of Acousticness by Tempo Category', y=1.03)
plt.show()
```
